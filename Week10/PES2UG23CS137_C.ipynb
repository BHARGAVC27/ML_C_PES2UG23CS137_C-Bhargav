## **Core Concepts**
**Support Vector Machine (SVM):** A powerful supervised learning algorithm that finds an optimal hyperplane to separate data points of different classes.

**Kernel Trick:** A technique that allows SVMs to solve non-linear problems by transforming data into a higher-dimensional space.

- Linear Kernel: Creates a straight-line decision boundary.

- RBF Kernel: Creates a complex, non-linear boundary, like a circle or a wave.

- Polynomial Kernel: Creates a curved, polynomial decision boundary.

## 2. Helper Function for Visualization
This helper function plots the decision boundaries for our trained models. It will work for all our datasets

```python
def plot_decision_boundaries(X, y, model, title):
    """
    Visualizes the decision boundaries of a trained classifier.
    (Corrected version)
    """
    # Create a meshgrid to plot the decision boundary
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Predict the class for each point in the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary and the data points
    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

    # Plot the training points
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)

    # Get unique labels and ensure they are a list for the legend function
    unique_labels = np.unique(y)
    if len(unique_labels) == 2:
        legend_labels = ['Class 0', 'Class 1']
    else:
        legend_labels = list(unique_labels.astype(str)) # Convert numpy array to a list

    plt.legend(handles=scatter.legend_elements()[0], labels=legend_labels)
    plt.show()
```

## Dataset 1: The Moons Dataset
The Moons dataset is a synthetic dataset designed to test non-linear classification algorithms. The data points are shaped like two interlocking half-moons.

### Step 1.1: Generate and Prepare the Data
We will generate the data using scikit-learn and apply feature scaling.

```python
from sklearn.datasets import make_moons

# Generate non-linearly separable data
X_moons, y_moons = make_moons(n_samples=500, noise=0.2, random_state=42)

# Split the data into training and testing sets
X_train_moons, X_test_moons, y_train_moons, y_test_moons = train_test_split(
    X_moons, y_moons, test_size=0.3, random_state=42
)

# Scale the features
# Feature scaling is crucial for SVMs to perform optimally.
scaler_moons = StandardScaler()
X_train_moons_scaled = scaler_moons.fit_transform(X_train_moons)
X_test_moons_scaled = scaler_moons.transform(X_test_moons)
```

```python
# Visualize the Moons dataset
plt.figure(figsize=(8, 6))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Moons Dataset')
plt.show()
```

## Plot Interpretation  Moons scatter plot

This scatter plot shows the raw Moons dataset points colored by their true labels.

What to look for:
- Two crescent-shaped clusters (half-moons) representing the two classes. They interlock, creating a non-linear boundary between classes.
- Because of the `noise=0.2` parameter, some overlap exists  this creates misclassification risk for linear models.

Implication for SVM kernels:
- A linear SVM will struggle to find a straight line that cleanly separates the classes.
- Non-linear kernels (RBF, Polynomial) can map features into spaces where separation is easier, producing curved boundaries that follow the moon shapes.

### Step 1.2: Train and Evaluate SVM Kernels

```python
kernels = ['linear', 'rbf', 'poly']
models_moons = {}

for kernel in kernels:
    # Initialize SVM with the current kernel
    if kernel == 'poly':
        svm_model = SVC(kernel=kernel, degree=3, random_state=42)
    else:
        svm_model = SVC(kernel=kernel, random_state=42)

    # Train the model on scaled training data
    svm_model.fit(X_train_moons_scaled, y_train_moons)

    # Store the trained model
    models_moons[kernel] = svm_model

    # Make predictions on the scaled test set
    y_pred_moons = svm_model.predict(X_test_moons_scaled)

    # Print classification report with SRN
    print(f"SVM with {kernel.upper()} Kernel <PES2UG23CS137>")
    print(classification_report(y_test_moons, y_pred_moons))
    print("-" * 40 + "\n")
```

## Explanation  Training and evaluating SVMs on Moons

This cell trains three SVM classifiers (linear, RBF, polynomial) on the scaled Moons training data and evaluates them on the scaled test data.

What the code does:
- Iterates over kernels = ['linear', 'rbf', 'poly'].
- Initializes SVC: `SVC(kernel=..., random_state=42)`; polynomial uses degree=3 for a simple polynomial mapping.
- Fits the model on `X_train_moons_scaled` and `y_train_moons`.
- Predicts labels for `X_test_moons_scaled` and prints a classification report for each kernel.

Why this matters:
- The `classification_report` contains precision, recall, and F1-score for each class. For non-linear data like Moons, expect the RBF and polynomial kernels to have higher F1-scores than the linear kernel.

Inputs: scaled train/test arrays and kernel list.
Outputs: trained models stored in `models_moons` and printed classification reports for each kernel.

### Step 1.3: Visualize Decision Boundaries

```python
#TODO: Replace with your SRN
for kernel, model in models_moons.items():
    plot_decision_boundaries(
        X_train_moons_scaled,
        y_train_moons,
        model,
        title=f'Moons Dataset - SVM with {kernel.upper()} Kernel <PES2UG23CS137>'
    )
```

## Plot Interpretation  Moons decision boundaries

This cell iterates through the trained `models_moons` and calls `plot_decision_boundaries` to visualize each kernel's decision frontier on the training data.

How to interpret the three plots:

- Linear kernel plot:
  - Expect a straight-ish boundary (a linear separator). Because the Moons dataset is non-linear, the linear SVM likely misclassifies many points near the overlapping regions.

- RBF kernel plot:
  - RBF (Gaussian) can form complex, curved boundaries that wrap around clusters. For the Moons dataset it usually follows the moon shapes better, producing fewer misclassified points.

- Polynomial kernel plot:
  - Polynomial of degree 3 can also create curved boundaries. Its flexibility depends on degree; degree=3 is often sufficient to capture some curvature but might under/overfit depending on data/noise.

## 4. Understanding the Hard and Soft Margins
Soft Margin (Small C value, e.g., 0.1): A smaller C value creates a "softer" margin, meaning the model is more tolerant of misclassifications. This results in a wider margin and can lead to better generalization, especially with noisy data.

Hard Margin (Large C value, e.g., 100): A larger C value creates a "harder" margin. The model will try to classify every data point correctly, resulting in a narrower margin. This can lead to overfitting if the data has outliers.

Let's create a dataset that is mostly linearly separable but has some noise, which is perfect for understanding the difference between hard and soft margins.

```python
from sklearn.datasets import make_blobs

# Generate linearly separable data with some noise
X_linear, y_linear = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=0.60)

# Add some outliers
outliers_X = np.array([[0.5, 2.5], [1.5, 0.5]])
outliers_y = np.array([1, 0])
X_linear = np.concatenate([X_linear, outliers_X])
y_linear = np.concatenate([y_linear, outliers_y])


# Split and scale the data
X_train_linear, X_test_linear, y_train_linear, y_test_linear = train_test_split(
    X_linear, y_linear, test_size=0.3, random_state=42
)
scaler_linear = StandardScaler()
X_train_linear_scaled = scaler_linear.fit_transform(X_train_linear)
X_test_linear_scaled = scaler_linear.transform(X_test_linear)
```

Now, let's train two SVM models with a linear kernel: one with a small C (soft margin) and one with a large C (hard margin).

```python
# Soft Margin SVM (small C)
# Create a linear SVM model with a small C value (soft margin)
svm_soft = SVC(kernel='linear', C=0.1, random_state=42)

# Fit the soft margin model to the training data
svm_soft.fit(X_train_linear_scaled, y_train_linear)

# Plot with SRN in the title
plot_decision_boundaries(X_train_linear_scaled, y_train_linear, svm_soft, title='Soft Margin SVM (C=0.1) <PES2UG23CS137>')

# Hard Margin SVM (large C)
# Create a linear SVM model with a large C value (hard margin)
svm_hard = SVC(kernel='linear', C=100, random_state=42)

# Fit the hard margin model to the training data
svm_hard.fit(X_train_linear_scaled, y_train_linear)

# Plot with SRN in the title
plot_decision_boundaries(X_train_linear_scaled, y_train_linear, svm_hard, title='Hard Margin SVM (C=100) <PES2UG23CS137>')
```

## Plot Interpretation  Soft vs Hard Margin (Linear SVM)

This cell trains two linear SVMs: a soft-margin model (C=0.1) and a hard-margin model (C=100), then visualizes their decision boundaries.

How to interpret the plots:
- Soft margin (C=0.1):
  - The model allows some misclassifications to obtain a wider margin (less sensitive to outliers). You will likely see a smoother, wider boundary that ignores the outliers.
- Hard margin (C=100):
  - The model penalizes misclassifications heavily, trying to correctly classify every point even at the cost of a narrow margin. It may bend the boundary to correctly classify outliers and thus overfit.
